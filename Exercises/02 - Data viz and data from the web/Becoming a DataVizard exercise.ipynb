## Visualizing two continuous variables

- Let's see if blockbusters with higher ratings also make more revenue!

```python
plt.scatter(movies['worldwide_gross'], movies['imdb_rating'])

plt.xlabel('Worldwide gross revenue')
plt.ylabel('IMDB rating')
```

```python
plt.scatter(movies['worldwide_gross'], movies['imdb_rating'], s = 2)

plt.xlabel('Worldwide gross revenue')
plt.ylabel('IMDB rating')
```

```python
# Calculate average yearly worldwide gross revenue
years_stats = movies.groupby('year')['worldwide_gross'].agg(['mean', 'std'])
years_stats
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming avg_yearly_worldwide_gross is the DataFrame from your groupby operation.
# For demonstration, let's calculate a random error (replace it with actual error measure if available).

# Plot with error bars
plt.errorbar(years_stats.index, 
             years_stats['mean'], 
             yerr=years_stats['std'], capsize=5, label='Average Worldwide Gross')

# Customize plot
plt.xlabel('Year')
plt.ylabel('Average Worldwide Gross ($)')
plt.title('Average Yearly Worldwide Gross with Error Bars')
plt.legend()

plt.show()
```

```python
# Write your code to make filled plot here
plt.fill_between(years_stats.index, years_stats['mean'] - years_stats['std'], years_stats['mean'] + years_stats['std'], alpha = 0.5)

plt.plot(years_stats.index, years_stats['mean'])

plt.xlabel('Year')
plt.ylabel('Worldwide gross revenue')

plt.show()
```

```python
# Calculate stats by genre here
# for genre in movies['Main_Genre'].unique():
#     print(genre)

movies.groupby('Main_Genre')['length']
```

```python
# Group by 'Main_Genre' and get movie lengths
grouped = movies.groupby('Main_Genre')['length']

# Set up a 4x4 grid for the subplots
fig, axes = plt.subplots(4, 4, figsize=(16, 16))

# Flatten the axes array to easily loop over
axes = axes.flatten()

# Common x and y range for all plots (adjust based on your data)
common_x_range = (movies['length'].min(), movies['length'].max())
common_y_range = (0, movies['length'].value_counts().max())

# Loop over each genre and corresponding axis
for i, (genre, lengths) in enumerate(grouped):
    # Plot histogram for the current genre
    axes[i].hist(lengths, bins=20, alpha=0.7)
    
    # Set title and limits to ensure uniform scale
    axes[i].set_title(genre)
    axes[i].set_xlim(common_x_range)
    axes[i].set_ylim(common_y_range)

# Add x and y axis labels outside the grid
fig.text(0.5, 0.01, 'Movie Length', ha='center', fontsize=16)
fig.text(0.01, 0.5, 'Frequency', va='center', rotation='vertical', fontsize=16)


plt.show()
```

Most movies are around 2h long. Romance movies tend to be longer and Amimation shorter.

A common way to represent [power-laws](https://en.wikipedia.org/wiki/Power_law) is to use a histogram using a log-log scale  -- remember: the x-axis of an histogram is segmented in bins of equal sizes and y-values are the average of each bin. As shown in Fig.(b), small bins sizes might introduce artifacts. Fig.(b) and Fig.(c) are examples of histograms with two different bin sizes. Another great way to visualize such distribution is to use a cumulative representation, as shown in Fig.(d), in which the y-axis represents the number of data points with values greater than y. Create the figures (b) and (d) using the data extracted for task 1A. For Fig.(b), represent two histograms using two different bin sizes and provide a brief description of the results. What does this tell you about the relationship between the two variables, namely the frequency of each bin and the value (i.e. population in case of the communal data) for each bin?

<img src="plaw.png" style="width: 600px;" align="left">
  
The figure is extracted from [this paper](https://arxiv.org/pdf/cond-mat/0412004.pdf) that contains more information about this family of distributions.

```python
# Power law with a lot of bins
pop_per_commune['population_Dec'].plot(kind='hist', loglog=True, bins=10000,histtype='step')
plt.show()
# plt.xscale('log')
# plt.yscale('log')
```

```python
# Power law with fewer bin
pop_per_commune['population_Dec'].plot(kind='hist', loglog=True, bins=100,histtype='step')
plt.show()
```

```python
# CCDF
pop_per_commune['population_Dec'].plot(kind='hist', log=True, bins=100, histtype='step', cumulative=-1)
plt.show()
```

```python
# Generate the cumulative histogram data
array_cumulative = plt.hist(pop_per_commune.population_Dec, bins=100, log=True, cumulative=-1, histtype='step')

# Extract the bin edges and cumulative counts from the histogram
counts = array_cumulative[0]
bin_edges = array_cumulative[1]

# Plot using log-log scaling
plt.figure(figsize=(10, 6))
plt.loglog(bin_edges[:-1], counts, drawstyle='steps-post')  # Use the left edges of the bins for plotting

# Add labels and title
plt.title('Log-Log Cumulative Histogram of Population Counts', fontsize=16)
plt.xlabel('Population (log scale)', fontsize=14)
plt.ylabel('Cumulative Number of Communes (log scale)', fontsize=14)

plt.tight_layout()
plt.show()
```
